```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def convertToRGB(img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

#Nhận diện đôi mắt trước

```python
#load test image
test1 = cv2.imread('face_glasses_1.jpg')

#convert the test image to gray image as opencv face detector expects gray images
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
#load cascade classifier training file for haarcascade
haar_eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
eyes = haar_eye_cascade.detectMultiScale(gray_img,
                                           scaleFactor=1.01,
                                           minNeighbors=10)
```

```python
for (x, y, w, h) in eyes:
  cv2.rectangle(test1, (x, y), (x+w, y+h), (255, 0, 255), 2)
```

```python
#convert to RGB and show
plt.imshow(convertToRGB(test1))
```

```python
#Lưu lại hình ảnh
plt.imsave('test_02.jpg', convertToRGB(test1))
```

#Sau khi nhận diện đôi mắt, nhận diện khuôn mặt

```python
#load test image
test1 = cv2.imread('test_02.jpg')
#convert the test image to gray image as opencv face detector expects gray images
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
#load cascade classifier training file for haarcascade
haar_face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img,
                                           scaleFactor=1.1,
                                           minNeighbors=5)
```

```python
for (x, y, w, h) in faces:
  cv2.rectangle(test1, (x, y), (x+w, y+h), (255, 0, 255), 2)
```

```python
#convert to RGB and show
plt.imshow(convertToRGB(test1))
```

```python
#Lưu lại hình ảnh
plt.imsave('face_glasses_1_output.jpg', convertToRGB(test1))
```
